#### Modeling functions

```python
def nested_model(tdf, proteins, label, metric="auc", NUM_TRIALS=20):
    X = tdf[proteins]
    y = tdf[label]
    
    p_grid = {"svc__C": [1, 10, 100], "svc__gamma": [0.01, 0.1]}
    
    # Define pipeline: linear imputer + SVM
    pipeline = Pipeline([
        ("imputer",KNNImputer(n_neighbors=4)),
        ("svc", SVC(kernel="linear"))
    ])
    
    non_nested_scores = np.zeros(NUM_TRIALS)
    nested_scores = np.zeros(NUM_TRIALS)
    
    # Loop for each trial
    for i in range(NUM_TRIALS):
        inner_cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=i)
        outer_cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=i)

        if metric == "auc":
            # Non-nested parameter search and scoring
            clf = GridSearchCV(estimator=pipeline, param_grid=p_grid, cv=outer_cv, scoring='roc_auc')
            clf.fit(X, y)
            non_nested_scores[i] = clf.best_score_
        
            # Nested CV with parameter optimization
            clf = GridSearchCV(estimator=pipeline, param_grid=p_grid, cv=inner_cv, scoring='roc_auc')
            nested_score = cross_val_score(clf, X=X, y=y, cv=outer_cv)
            nested_scores[i] = nested_score.mean()
        if metric == "acc":
            # Non-nested parameter search and scoring
            clf = GridSearchCV(estimator=pipeline, param_grid=p_grid, cv=outer_cv)
            clf.fit(X, y)
            non_nested_scores[i] = clf.best_score_
        
            # Nested CV with parameter optimization
            clf = GridSearchCV(estimator=pipeline, param_grid=p_grid, cv=inner_cv)
            nested_score = cross_val_score(clf, X=X, y=y, cv=outer_cv)
            nested_scores[i] = nested_score.mean()
            
    score_difference = non_nested_scores - nested_scores
    
    print(
        "Average difference of {:6f} with std. dev. of {:6f}.".format(
            score_difference.mean(), score_difference.std()
        )
    )
    
    # Plot scores on each trial for nested and non-nested CV
    plt.figure()
    plt.subplot(211)
    (non_nested_scores_line,) = plt.plot(non_nested_scores, color="r")
    (nested_line,) = plt.plot(nested_scores, color="b")
    plt.hlines(y=[non_nested_scores.mean(), nested_scores.mean()], colors=['r', 'b'], linestyles=['--', '--'], xmin=0, xmax=NUM_TRIALS-1)
    print("non- nested average: ", non_nested_scores.mean(),"Nested average: ", nested_scores.mean())
    
    if metric == "auc":
        plt.ylabel("ROC AUC", fontsize="14")
    if metric == "acc":
        plt.ylabel("Accuracy (%)", fontsize="14")
        
    plt.legend(
        [non_nested_scores_line, nested_line],
        ["Non-Nested CV", "Nested CV"],
        bbox_to_anchor=(0, 0.4, 0.5, 0),
    )
    plt.title(
        "Non-Nested and Nested Cross Validation",
        x=0.5,
        y=1.1,
        fontsize="15",
    )
    
    # Plot bar chart of the difference.
    plt.subplot(212)
    difference_plot = plt.bar(range(NUM_TRIALS), score_difference)
    plt.xlabel("Individual Trial #")
    if metric == "auc":
        plt.ylabel("ROC AUC difference", fontsize="14")
    if metric == "acc":
        plt.ylabel("Accuracy difference", fontsize="14")
 
    plt.show()
```

```python
def bar_plot(tdf, proteins, label, NUM_TRIALS=20):
    X = tdf[proteins]
    y = tdf[label]
    
    p_grid = {"svc__C": [1, 10, 100], "svc__gamma": [0.01, 0.1]}
    
    # Define pipeline: linear imputer + SVM
    pipeline = Pipeline([
        ("imputer",KNNImputer(n_neighbors=4)),
        ("svc", SVC(kernel="linear"))
    ])
    
    non_nested_scores_auc = np.zeros(NUM_TRIALS)
    nested_scores_auc = np.zeros(NUM_TRIALS)
    non_nested_scores_acc = np.zeros(NUM_TRIALS)
    nested_scores_acc = np.zeros(NUM_TRIALS)
    
    # Loop for each trial
    for i in range(NUM_TRIALS):
        inner_cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=i)
        outer_cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=i)

        #AUC
        # Non-nested parameter search and scoring
        clf = GridSearchCV(estimator=pipeline, param_grid=p_grid, cv=outer_cv, scoring='roc_auc')
        clf.fit(X, y)
        non_nested_scores_auc[i] = clf.best_score_
    
        # Nested CV with parameter optimization
        clf = GridSearchCV(estimator=pipeline, param_grid=p_grid, cv=inner_cv, scoring='roc_auc')
        nested_score_auc = cross_val_score(clf, X=X, y=y, cv=outer_cv)
        nested_scores_auc[i] = nested_score_auc.mean()

        #ACCURACY
        # Non-nested parameter search and scoring
        clf = GridSearchCV(estimator=pipeline, param_grid=p_grid, cv=outer_cv)
        clf.fit(X, y)
        non_nested_scores_acc[i] = clf.best_score_
    
        # Nested CV with parameter optimization
        clf = GridSearchCV(estimator=pipeline, param_grid=p_grid, cv=inner_cv)
        nested_score_acc = cross_val_score(clf, X=X, y=y, cv=outer_cv)
        nested_scores_acc[i] = nested_score_acc.mean()

    data = [non_nested_scores_auc, nested_scores_auc, non_nested_scores_acc, nested_scores_acc]
    labels = ['non-nested AUC', 'nested AUC', 'non-nested accuracy', 'nested accuracy']
    
    # Compute means and standard deviations
    means = [np.mean(group) for group in data]
    stds = [np.std(group) for group in data]

    for i in range(len(labels)):
        print(labels[i], means[i], stds[i])
    
    plt.figure(figsize=(8, 6))
    x = np.arange(len(data))
    
    plt.bar(x, means, yerr=stds, capsize=8)
    plt.ylim(0.5, 1)
    
    plt.title('Accuracy and ROC AUC scores', fontsize=14)
    plt.ylabel('Score', fontsize=12)
    plt.xlabel('Methods', fontsize=12)
    plt.xticks(x, labels)
```

```python
def feature_importance(tdf, proteins, label, n_features_to_select = 1):
    # Load example dataset
    X = tdf[proteins]
    y = tdf[label]
    
    # Define the SVM model (use linear kernel for feature ranking)
    svm = SVC(kernel='linear', random_state=42)
    
    rfe = RFE(estimator=svm, n_features_to_select=n_features_to_select)
    
    # Apply RFE to the dataset
    X_rfe = rfe.fit_transform(X, y)
    
    print(f"Features: {X.columns}")
    print(f"Selected features (mask): {rfe.support_}")
    print(f"Ranking of features: {rfe.ranking_}")
    
    # Now, evaluate the model using StratifiedKFold with cross-validation
    cv = StratifiedKFold(n_splits=4, shuffle=True, random_state=42)
    
    # Because SVC is already used as estimator in RFE, 
    # train a new SVC on the selected features:
    svm_final = SVC(kernel='linear', random_state=42)
    
    # Cross-validate on the reduced feature set
    scores = cross_val_score(svm_final, X_rfe, y, cv=cv, scoring='accuracy')
    
    print(f"Cross-validated accuracy scores: {scores}")
    print(f"Mean accuracy: {scores.mean():.4f}")
```

```python
def svm_plot(tdf, proteins, label, control, case, kernel="linear", x_min=7, x_max=13, y_min=7, y_max=14):

    X = tdf[proteins].to_numpy()
    y = tdf[label]
    mapping = {case : 1, control : 0}
    y = [mapping[i] for i in y]
    y = np.array(y)
    
    long_title=True
    support_vectors=True
    ax=None
    
    # Train the SVC
    clf = SVC(kernel=kernel, C= 10, class_weight = 'balanced', gamma= 0.1).fit(X, y)

    # Classification report
    y_pred_train = clf.predict(X)
    report = classification_report(y, y_pred_train)
    print("AUC: ", metrics.roc_auc_score(y, y_pred_train))
    print(report)
    
    # Settings for plotting
    if ax is None:
        _, ax = plt.subplots(figsize=(4, 3))
    x_min, x_max, y_min, y_max =  x_min, x_max, y_min, y_max
    ax.set(xlim=(x_min, x_max), ylim=(y_min, y_max))
    
    # Plot decision boundary and margins
    common_params = {"estimator": clf, "X": X, "ax": ax}
    DecisionBoundaryDisplay.from_estimator(
        **common_params,
        response_method="predict",
        plot_method="pcolormesh",
        alpha=0.3,
    )
    DecisionBoundaryDisplay.from_estimator(
        **common_params,
        response_method="decision_function",
        plot_method="contour",
        levels=[-1, 0, 1],
        colors=["k", "k", "k"],
        linestyles=["--", "-", "--"],
    )
    
    if support_vectors:
        # Plot bigger circles around samples that serve as support vectors
        ax.scatter(
            clf.support_vectors_[:, 0],
            clf.support_vectors_[:, 1],
            s=150,
            facecolors="none",
            edgecolors="k",
        )
    
    # Plot samples by color and add legend
    s = ax.scatter(X[:, 0], X[:, 1], c=y, s=30, edgecolors="k")
    ax.legend(*s.legend_elements(), loc="upper right", title="Classes")
    if long_title:
        ax.set_title(f" Decision boundaries of {kernel} kernel in SVC")
    else:
        ax.set_title(kernel)
        
    ax.set_xlabel(proteins[0])
    ax.set_ylabel(proteins[1])
    
    if ax is None:
        plt.show()
```
