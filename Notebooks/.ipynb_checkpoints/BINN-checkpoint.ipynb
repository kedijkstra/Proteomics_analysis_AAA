```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from statsmodels.stats.multitest import multipletests
from scipy.stats import fisher_exact
import statsmodels.api as sm
import scipy.stats as ss
from sklearn.metrics import silhouette_score
from sklearn.linear_model import LogisticRegression
from matplotlib_venn import venn3
from imblearn.over_sampling import SMOTE
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.datasets import  make_classification
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, f1_score, accuracy_score
from sklearn.inspection import permutation_importance
import shap
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from itertools import combinations
import seaborn as sns

import sys
sys.path.append("..")
from Pipelines.DAP import DifferentiallyAbundantProteins as DAP

#Read datafile
proteomics_data = pd.read_csv("C:/Users/keimp/AAA_AMC/data/20250331 AAA plasma proteomics raw data.csv", delimiter=";")
#Rotate data into wide format
proteomics_data = proteomics_data.transpose()

#Convert the samples into a column
proteomics_data = proteomics_data.reset_index()
#Make the first row the column names
proteomics_data.columns = proteomics_data.iloc[0]
proteomics_data = proteomics_data[1:]
#rename sample column
proteomics_data = proteomics_data.rename(columns={'Genes': 'Sample_index'})
#Get the proteins
proteins = proteomics_data.columns.values.tolist()[1:]

meta_data = pd.read_csv("C:/Users/keimp/AAA_AMC/data/plasma_proteomics_meta_file_rawDataCSV.csv", delimiter=";")
df = pd.merge(proteomics_data, meta_data, on="Sample_index", how="inner")
df = df.loc[(df["Pathology"]=="AAA") | (df["Pathology"]=="Controle")]

for p in proteins:
    df[p] = df[p].astype(str).str.replace(',', '.')
    df[p] = pd.to_numeric(df[p], errors='coerce')

df = df.dropna(axis=1, how="all")
proteins.remove("FABP1")

df = df.reset_index(drop=True)
```

## Biologically informend neural network

### Data prep

```python
### TEMPORARY
d = DAP(df, "Sample_index", "SimplePathology", proteins)
d.missing_filtering(threshold=25, percentage=True)
df = d.get_data()[d.get_proteins()+["Sample_index","SimplePathology"]]
```

```python
#Randomly shuffle data
df = df.sample(frac=1).reset_index(drop=True)
```

```python
#CONVERT GENE SYMBOLS TO UNIPROT
from unipressed import IdMappingClient
import time
request = IdMappingClient.submit(
    source="GeneCards", dest="UniProtKB", ids=proteins
)

uniprot = {}
for p in list(request.each_result()):
    uniprot[p["from"]] = p["to"]

df = df.rename(columns=uniprot)
```

```python
class_0 = df[df["SimplePathology"] == "Control"]
class_1 = df[df["SimplePathology"] == "AAA"]

n = min(len(class_0), len(class_1))

# Sample each class
class_0_sampled = class_0.sample(n=n, random_state=42)
class_1_sampled = class_1.sample(n=n, random_state=42)

# Combine and shuffle
df = pd.concat([class_0_sampled, class_1_sampled]).sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
#DESIGN MATRIX
dm = df[["Sample_index","SimplePathology"]]

mapping = {"AAA" : 1, "Control" : 0}
dm["SimplePathology"] = [mapping[i] for i in list(dm["SimplePathology"])]
```

```python
#DATA MATRIX
t = [i for i in list(uniprot.values()) if i in list(df.columns)]
datam = df[["Sample_index"]+t]
datam = datam.transpose()

datam = datam.reset_index()

datam.columns = datam.iloc[0]
datam = datam.drop(datam.index[0])

datam = datam.rename(columns={"Sample_index":"Protein"})
```

### Train model

```python
from binn import BINN, BINNDataLoader, BINNTrainer, BINNExplainer
# Initialize BINN
binn = BINN(data_matrix=datam, network_source="reactome", n_layers=2
            , dropout=0)

# Initialize DataLoader
binn_dataloader = BINNDataLoader(binn)

# Create DataLoaders
dataloaders = binn_dataloader.create_dataloaders(
    data_matrix=datam,
    design_matrix=dm,
    feature_column="Protein",
    group_column="SimplePathology",
    sample_column="Sample_index",
    batch_size=78,
    validation_split=0.2,
)
# Train the model
trainer = BINNTrainer(binn)
trainer.fit(dataloaders=dataloaders, num_epochs=3)
```

```python
# Explain the model
explainer = BINNExplainer(binn)
explanations = explainer.explain_single(
    dataloaders
)
explanations
```

```python
from binn.plot.network import visualize_binn

layer_specific_top_n = {"0": 5, "1": 5, "2": 5, "3": 5, "4": 5, "5": 5}
plt = visualize_binn(
    explanations,
    top_n=layer_specific_top_n,
    plot_size=(20, 15),
    sink_node_size=500,
    node_size_scaling=200,
    edge_width=1,
    node_cmap="coolwarm",
    pathways_mapping="reactome",
    input_entity_mapping="uniprot"
)
plt.title("Interpreted network")
```

```python
print(trainer.evaluate(dataloaders["train"]))
print(trainer.evaluate(dataloaders["val"]))
```

## Grid search

```python
def build_model(data_matrix, design_matrix, drop_out, n_layers, n_epochs, 
               feature_col, group_col, sample_col, batch_size, val_split=0.2):
    # Initialize BINN
    binn = BINN(data_matrix=data_matrix, network_source="reactome", n_layers=n_layers
                , dropout=drop_out)
    
    # Initialize DataLoader
    binn_dataloader = BINNDataLoader(binn)
    
    # Create DataLoaders
    dataloaders = binn_dataloader.create_dataloaders(
        data_matrix=data_matrix,
        design_matrix=design_matrix,
        feature_column=feature_col,
        group_column=group_col,
        sample_column=sample_col,
        batch_size=batch_size,
        validation_split=val_split
    )
    # Train the model
    trainer = BINNTrainer(binn)
    trainer.fit(dataloaders=dataloaders, num_epochs=n_epochs)
    return trainer.evaluate(dataloaders["train"]), trainer.evaluate(dataloaders["val"]), binn, dataloaders
```

```python
#Search 
layers = list(range(1, 10))
drop_outs = list(range(0, 11, 2))
drop_outs = [i/10 for i in drop_outs]
#max_waiting = 10
max_epoch = 50

results = {}

x = 0
for l in layers:
    for d in drop_outs:
        print("Layer: ", l, " drop out: ", d, " n: ", x)
        train, val, _, _ = build_model(datam, dm, d, l, max_epoch, 
           "Protein", "SimplePathology", "Sample_index", 78)
        results[x] = [train["loss"], train["accuracy"] ,val["loss"], val["accuracy"]]
        x+=1
```

```python
results
```

```python
x=-1
while x < 10:
    x+=1
    train, val, binn, dataloaders = build_model(datam, dm, 0.4, 6, max_epoch, 
               "Protein", "SimplePathology", "Sample_index", 78)
    print([train["loss"], train["accuracy"] ,val["loss"], val["accuracy"]])
    
    explainer = BINNExplainer(binn)
    explanations = explainer.explain_single(
        dataloaders
    )
    
    layer_specific_top_n = {"0": 5, "1": 5, "2": 5, "3": 5, "4": 5, "5": 5}
    plt = visualize_binn(
        explanations,
        top_n=layer_specific_top_n,
        plot_size=(20, 15),
        sink_node_size=500,
        node_size_scaling=200,
        edge_width=1,
        node_cmap="coolwarm",
        pathways_mapping="reactome",
        input_entity_mapping="uniprot"
    )
    plt.title("Interpreted network")
    plt.show()
```
