## EDA of AAA meta data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from collections import Counter
```

### Load data

```python
meta_data = pd.read_csv("C:/Users/keimp/AAA_AMC/data/plasma_proteomics_meta_file_rawDataCSV.csv", delimiter=";")
meta_data.head()
```

```python
#Remove controls
meta_data = meta_data[meta_data.SimplePathology != "Control"]
#Check results
print(meta_data["SimplePathology"].unique())
```

#### Missing values

```python
cols = ['Pathology', 'SimplePathology', 'Sex',
       'Age', 'BMI', 'Size_AA', 'Smoking', 'Years_to_quitting', 'Diabetes',
       'DM_meds', 'Hypertension', 'HyperCholstero', 'Previous_vascu_surgery',
       'Renal_dysfunction', 'Clinical presentation',
       'Angio related familial history', 'Post OP growth (WIP)',
       'Type Aneurysms']
data = []
for c in cols:
    data.append(meta_data[c].isna().sum())

fig, ax = plt.subplots()
ax.bar(cols, data, color=["#2ca02c"])

ax.set_ylabel("Number of missing values")
ax.set_title("Missing values per column")
plt.xticks(rotation=90)
plt.show()
```

#### BMI

### Correlations

#### BMI and AA size

```python
bmi = list(meta_data["BMI"])
aa_size = list(meta_data["Size_AA"])
sex = list(meta_data["Sex"])
bmi_m = []
aa_size_m = []
bmi_f = []
aa_size_f = []

for n in range(len(bmi)):
    if str(bmi[n]) != "nan" and str(aa_size[n]) != "nan":
        if sex[n] == "M":
            bmi_m.append(float(str(bmi[n]).replace(',', '.')))
            aa_size_m.append(float(str(aa_size[n]).replace(',', '.')))
        if sex[n] == "F":
            bmi_f.append(float(str(bmi[n]).replace(',', '.')))
            aa_size_f.append(float(str(aa_size[n]).replace(',', '.')))

fig, ax = plt.subplots()
ax.scatter(aa_size_m, bmi_m, c=["#2ca02c"], label="Male")
ax.scatter(aa_size_f, bmi_f, c=["#d62728"], label="Female")
ax.set_title("BMI and AA size")
ax.set_xlabel("AA size (mm)")
ax.set_ylabel("body mass index (Kg/M2)")
ax.legend()
plt.show()
```

#### Smoking and AA size

```python
smoking = Counter([x.split(" ")[0] for x in meta_data["Smoking"]])
data = []

for i in smoking.keys():
    data.append([float(str(j).replace(',', '.')) for j in meta_data[meta_data.Smoking.str.startswith(i)]["Size_AA"] if str(j) != "nan"])

fig, ax = plt.subplots()
bplot = ax.boxplot(data, patch_artist=True)
ax.set_xticklabels(smoking.keys())
ax.set_title("Smoking status and AA size")
ax.set_ylabel("AA size (mm)")

for patch, color in zip(bplot['boxes'], ["#2ca02c","#d62728", "#1f77b4","#9467bd"]):
    patch.set_facecolor(color)
plt.show()
```

#### Age and AA size

```python
age = list(meta_data["Age"])
aa_size = list(meta_data["Size_AA"])
age_ = []
aa_size_ = []

for n in range(len(aa_size)):
    if str(age[n]) != "nan" and str(aa_size[n]) != "nan":
        age_.append(float(str(age[n]).replace(',', '.')))
        aa_size_.append(float(str(aa_size[n]).replace(',', '.')))

fig, ax = plt.subplots()
ax.scatter(age_, aa_size_, c=["#2ca02c"])
ax.set_title("age and AA size")
ax.set_ylabel("AA size (mm)")
ax.set_xlabel("age (years)")

b, a = np.polyfit(age_, aa_size_, deg=1)
xseq = np.linspace(40, 90, num=100)
ax.plot(xseq, a + b * xseq, color="k", lw=2.5)

plt.show()
```
